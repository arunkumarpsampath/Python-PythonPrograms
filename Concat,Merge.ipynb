# Normal distribution

```python
np.random.seed(123)
norm_dist = pd.DataFrame(np.random.randn(1,4))
norm_dist
```

```python
#np.random.seed(124)
binom_dist = pd.DataFrame(np.random.binomial(100,0.5,(1,4)))
binom_dist
```

```python
#np.random.seed(124)
poisson_dist = pd.DataFrame(np.random.poisson(100,(1,4)))
poisson_dist
```

```python
dist=[norm_dist, binom_dist, poisson_dist]
dist
```

```python
stat_dist = pd.concat(dist,keys=['Normal','Binomial','Poisson'])
stat_dist
```

```python
#first two rows and 3 cols
slice1 = stat_dist.iloc[:2,:]
slice1
```

```python
slice2 = stat_dist.iloc[2:]
slice2
```

```python
pd.concat([slice1,slice2],axis=0)
```

# Merge

```python
us_data = pd.read_csv("D:/Python_programming/Dataset/us_index_data.csv")
us_data = us_data.set_index('TradingDate')
us_data.head()
```

```python
slice1 = us_data.iloc[:2,:3]
slice1
```

```python
slice2 = us_data.iloc[:2,[3]]
slice2
```

```python
slice3 = us_data.iloc[1:3,:3]
slice3
```

```python
pd.merge(slice1,slice2)
```

```python
print('inner')
print(pd.merge(slice2,slice3,how='inner'))
print('\n')
print('outer')
print(pd.merge(slice2,slice3,how='outer'))
print('\n')
print('left')
print(pd.merge(slice2,slice3,how='left'))
print('\n')
print('right')
print(pd.merge(slice2,slice3,how='right'))
```

```python
slice1.join(slice2)
```
