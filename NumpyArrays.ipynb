```python
#Case study
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(12)

all_walks=[]
for k in range(10):
    random_walk=[0]
    step=0
    for i in range(100):
        dice = np.random.randint(1,7)
        
        
        if dice <= 2:
            step = max(0,step - 1)
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)
print(all_walks)
plt.plot(all_walks)
plt.show()
```
