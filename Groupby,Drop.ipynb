# Group by two indexes or multi index grouping

```python
euroleague.head()
```

```python
#to reset indexes
euroleague = euroleague.reset_index()
```

```python
euro_double_index = euroleague.set_index(['Month','Stat'])
euro_double_index.head()
```

```python
euro_double_grp = euro_double_index.groupby(level = ['Month','Stat'])
 
for name , group in euro_double_grp:
    print("\n")
    print(name)
    print(group)
```

```python
euro_double_stat = euro_double_index.groupby(level = ['Stat'])
euro_double_stat.sum()
```

```python
euroleague = euroleague.reset_index()
euroleague = euroleague.set_index('Month')
euroleague.head()
```

```python
euroleague.groupby('Month').sum()
```

```python
#euro_double_stat.loc['Total'] = 
euro_double_stat_sum = euro_double_stat.sum()
euro_double_stat_sum.head()
```

```python
euro_double_stat_sum.loc['Total'] =  euro_double_stat_sum.loc['GoalsScored'] / euro_double_stat_sum.loc['MatchesPlayed']
euro_double_stat_sum
```

```python
euro_double_stat_sum.sort_values(by = 'EPL')
```

```python
#alternate
grouping =  euro_double_stat_sum.loc['GoalsScored'] / euro_double_stat_sum.loc['MatchesPlayed']
grouping = pd.DataFrame(grouping).T
grouping = grouping.rename(index = {0 : 'Goals per Game'})
euro_total = pd.DataFrame.append(euro_double_stat_sum,grouping)
euro_total.head()
```

```python
#dropping a column
euro_total = euro_total.drop(['index'],axis = 1)
```

```python
#dropping a row
euro_total = euro_total.drop(['Goals per Game'],axis = 0)
print(euro_total)
print(grouping)
#appending the same row again
euro_total = pd.DataFrame.append(euro_total,grouping)
euro_total.head()
```

```python
euro_total = pd.DataFrame.append(euro_total,grouping)
euro_total.head()
```

```python
#to drop rows with duplicate index names
euro_total = euro_total.drop_duplicates()
euro_total.head()
```
